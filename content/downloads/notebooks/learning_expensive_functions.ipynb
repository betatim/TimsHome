Powerful features (or variables) are often expensive to compute. Maybe they require a lot of CPU power to compute or require human interaction. As a result the size of the datasets for which the features are evaluated are smaller than they should be.

This post explores a possible way around this. Regression algorithms can learn complex functions and are cheap to evaluate. The idea then is:

* Train a regression algorithm on a small subset of the data for which all expensive features have been calculated
* Use this algorithm to cheaply calculate the features for a much larger number of samples

Here I demonstrate this approach using a feature called "[Missing mass calculator](//arxiv.org/abs/1012.4686)". It is used in searches for the [Higgs boson](//en.wikipedia.org/wiki/Higgs_boson) and takes about one minute to calculate per sample (or event).

```python
from operator import itemgetter

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import StratifiedKFold, KFold
from sklearn.grid_search import RandomizedSearchCV
from scipy.stats import randint as sp_randint
from scipy.stats import uniform as sp_uniform
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor, RandomForestRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv('/tmp/atlas-higgs-challenge-2014-v2.csv.gz',
                 compression='gzip')
```

```python
print ', '.join(df.columns)
```

```python
columns = ['DER_mass_vis',
           'PRI_tau_pt', 'PRI_tau_eta', 'PRI_tau_phi', 'PRI_lep_pt',
           'PRI_lep_eta', 'PRI_lep_phi', 'PRI_met', 'PRI_met_phi', 'PRI_met_sumet']
columns += "DER_mass_transverse_met_lep, DER_mass_vis, DER_pt_h, DER_deltaeta_jet_jet, DER_mass_jet_jet, DER_prodeta_jet_jet, DER_deltar_tau_lep, DER_pt_tot, DER_sum_pt, DER_pt_ratio_lep_tau, DER_met_phi_centrality, DER_lep_eta_centrality".split(', ')

signal = df[(df.Label == 's') & (df.DER_mass_MMC > -800)]
X = signal[columns]
w = signal['Weight']
y = signal['DER_mass_MMC']

print X.shape
print w.shape
print y.shape
```

```python
X, w, y = X.as_matrix(), w.as_matrix(), y.as_matrix()
X_dev,X_eval, w_dev,w_eval, y_dev,y_eval = train_test_split(X, w, y, random_state=9548)
```

```python
print X_dev.shape
print type(X_dev)
```

```python
X_sub,_, y_sub,_ = train_test_split(X_dev, y_dev, train_size=2000, random_state=936)
sns.corrplot(np.c_[X_sub, y_sub])
```

```python
clf = RandomForestRegressor(n_estimators=40,
                            max_depth=3,
                            #max_features='sqrt',
                            random_state=235, n_jobs=7)
clf = GradientBoostingRegressor(n_estimators=1000,
                                #learning_rate=0.01,
                                subsample=0.4,
                                max_features=0.4,
                                max_depth=3)

scores = []
test_scores = []
train_scores = []
cv = KFold(X_dev.shape[0], 3)
for train,test in cv:
    X_train,y_train,w_train = X_dev[train], y_dev[train], w_dev[train]
    X_test,y_test,w_test = X_dev[test], y_dev[test], w_dev[test]
    
    clf.fit(X_train, y_train, sample_weight=w_train)
    y_pred_train = clf.predict(X_train)
    y_pred = clf.predict(X_test)
    scores.append(clf.score(X_test, y_test, sample_weight=w_test))
    print "train, test error:",
    print mean_squared_error(y_train, y_pred_train, w_train),
    print mean_squared_error(y_test, y_pred, w_test)
    
    test_score = np.zeros((clf.n_estimators,), dtype=np.float64)
    test_scores.append(test_score)
    for i, y_pred in enumerate(clf.staged_decision_function(X_test)):
        test_score[i] = mean_squared_error(y_test, y_pred, w_test)
        
    train_score = np.zeros((clf.n_estimators,), dtype=np.float64)
    train_scores.append(train_score)
    for i, y_pred in enumerate(clf.staged_decision_function(X_train)):
        train_score[i] = mean_squared_error(y_train, y_pred, w_train)

print scores
print np.mean(scores), np.std(scores)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)

for test,train in zip(test_scores, train_scores):
    plt.plot(np.arange(clf.n_estimators) + 1, train,
             'b-', alpha=0.5,
             label='Training Set MSE')
    plt.plot(np.arange(clf.n_estimators) + 1, test,
             'r-', alpha=0.5,
             label='Test Set MSE')
plt.xlabel("Boosting iteration")
plt.ylabel("MSE")
```

```python
# Utility function to report best scores
def report(grid_scores, n_top=3):
    top_scores = sorted(grid_scores, key=itemgetter(1), reverse=True)[:n_top]
    for i, score in enumerate(top_scores):
        print "Model with rank: {0}".format(i + 1)
        print "Mean validation score: {0:.3f} (std: {1:.3f})".format(
              score.mean_validation_score,
              np.std(score.cv_validation_scores))
        print "Parameters: {0}".format(score.parameters)
        print
        
n_iter_search = 20
clf = GradientBoostingRegressor(n_estimators=1000,
                                #learning_rate=0.01,
                                subsample=0.4,
                                max_features=0.4,
                                max_depth=3)
param_dist = {"n_estimators": [1000, 2000],
              "max_depth": [3, 5, 1, 7],
              "learning_rate": [0.5, 0.1, 0.05, 0.01],
              "max_features": sp_uniform(),
              "subsample": sp_uniform(),
              #"min_samples_split": sp_randint(1, 11),
              #"min_samples_leaf": sp_randint(1, 11),
              #"bootstrap": [True, False],
              "loss": ["ls", "lad"],
              }
random_search = RandomizedSearchCV(clf,
                                   param_distributions=param_dist,
                                   n_iter=n_iter_search,
                                   n_jobs=7)
random_search.fit(X_dev, y_dev)
```

```python
report(random_search.grid_scores_, 10)
```
